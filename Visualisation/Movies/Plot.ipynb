```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
svd_rf = np.array([0.3547619047619048,0.3607142857142857,0.35833333333333334])
svd_lr = np.array([0.2357142857142857,0.2369047619047619,0.2369047619047619])
ae_rf = np.array([0.09404761904761905,0.10833333333333334,0.0880952380952381])
ae_lr = np.array([0.08214285714285714,0.08452380952380953,0.10833333333333334])
full_rf = 0.4035714285714286
full_lr = 0.40714285714285714
```

```python
svd_rf = np.append(svd_rf, full_rf)
ae_rf = np.append(ae_rf, full_rf)

svd_lr = np.append(svd_lr, full_lr)
ae_lr = np.append(ae_lr, full_lr)
```

```python
xs = [1,2,3,4]
```

```python
plt.figure(figsize=(8,8))
plt.plot(xs, svd_lr, "r", label="svd, LR")
plt.plot(xs, svd_rf, 'm', label="svd, RF")
plt.plot(xs, ae_lr, 'b', label="ae, LR")
plt.plot(xs, ae_rf, 'c', label="ae, RF")
plt.grid()
plt.xticks(ticks=xs, labels=['50', '100', '500', 'full data'])
plt.xlabel("Number of features in the compressed data")
plt.ylabel("Accuracy of the classification algorithm")
plt.title("Classification accurasy vs. compression type and size")

plt.hlines(full_rf, xmin=1, xmax=4, linestyles='dashed', colors='b', label='full, RF')
plt.hlines(full_lr, xmin=1, xmax=4, linestyles='dashed', colors='r', label='full, LR')

plt.legend()
```
